```python
def find_dup(bow_matrix,tmp_indices):
    dup_dict = {}
    doc_num = np.shape(bow_matrix)[0]
    for i in range(doc_num-1):
        diff_matrix = np.abs(bow_matrix[i+1:,:] - bow_matrix[i])
        diff_array = np.sum(diff_matrix,axis = 1)
        #print(diff_array)
        dup_indices = np.where(diff_array<2)[0]
        if len(dup_indices):
            dup_dict[tmp_indices[i]] = np.array(tmp_indices)[i+dup_indices+1] 
    return dup_dict

```

```python
'''    print_num = 0#5
        print_count = 0
                if print_count<print_num:
                    print_count+=1
                    #import pdb; pdb.set_trace()
                    print(str(k) + ": is duplicated with " + str(v))
'''
def batch_dup_check(hadm_list,num,start):
    tmp_list = hadm_list[start: min(len(hadm_list),start + num)]
    to_delete = []
    for k in tqdm(tmp_list):
        tmp_df = target_df[target_df.HADM_ID == k]
        tmp_indices = tmp_df.index
        tmp_text_list = list(tmp_df.TEXT)
        bow_matrix = CountVectorizer(stop_words = 'english').fit_transform(tmp_text_list).todense()
        dup_dict = find_dup(bow_matrix,tmp_indices)
        if len(dup_dict):
            for k,dup_indices in dup_dict.items():
                for ind in dup_indices:
                    #import pdb; pdb.set_trace()
                    if not pd.to_datetime(tmp_df.loc[ind].CHARTDATE) == pd.to_datetime(tmp_df.loc[k].CHARTDATE):
                        to_delete.append(ind)
    return to_delete
```
